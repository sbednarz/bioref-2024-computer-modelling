```python

# y=x**2-2
# y=2x
# 

from scipy.optimize import fsolve
import numpy as np








def system(X):
    x = X[0] # unpack x
    y = X[1] # unpack y
    
    eq1 = y - x**2 + 2
    eq2 = y - 2*x
    return [eq1, eq2]




























guess = [-10, -10]

x0, y0 = fsolve(system, guess)
print("x0, y0 = {} {}".format(x0, y0))





























# x = -0.73 y = -1.46
# x = 2.73 y = 5.46
```
