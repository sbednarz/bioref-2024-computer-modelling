```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint


# Develop a (simplified) model of polymerization:

# M1 + M1 => M2, kp
# M2 + M1 => M3, kp
# M3 + M1 => M4, kp
# M4 + M1 => M5, kp

# Validate the model (mass balance).
# When the monomer dissapear?
# When pentamers appear in the reaction mixture?
# Plot evolution of the system in time.

# M0 = 0.1 mol/L
# kp = 1e-3 L/mol*s
```

```python
def model(y, t):
  M1 = y[0]
  M2 = y[1]
  M3 = y[2]
  M4 = y[3]
  M5 = y[4]

  dM1dt = -2*kp*M1*M1 - kp*M2*M1 - kp*M3*M1 - kp*M4*M1
  dM2dt = kp*M1*M1  -  kp*M2*M1
  dM3dt = kp*M2*M1  -  kp*M3*M1
  dM4dt = kp*M3*M1  -  kp*M4*M1
  dM5dt = kp*M4*M1

  return [dM1dt, dM2dt, dM3dt, dM4dt, dM5dt]

M0 = 0.1
kp = 1e-1
ic = [M0, 0, 0, 0, 0]

t = np.linspace(0,1000)

results = odeint(model, ic, t)
M1 = results[:,0]
M2 = results[:,1]
M3 = results[:,2]
M4 = results[:,3]
M5 = results[:,4]

plt.plot(t, M1, label='M1')
plt.plot(t, M2, label='M2')
plt.plot(t, M3, label='M3')
plt.plot(t, M4, label='M4')
plt.plot(t, M5, label='M5')
plt.legend()

M1_ = M1[-1]
M2_ = M2[-1]
M3_ = M3[-1]
M4_ = M4[-1]
M5_ = M5[-1]


print(M1 + 2*M2 + 3*M3 + 4*M4 + 5*M5)
```
