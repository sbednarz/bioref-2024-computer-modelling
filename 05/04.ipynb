```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

# T neq const
# A -> B, k
# k = k0 exp(-Ea/RT)

def model(y, t):
    A = y[0]
    B = y[1]
    T = y[2]
    k = k0*np.exp(-Ea/(R*T)) # calculate instant k at T
    dAdt = -k*A
    dBdt = k*A    
    dTdt = rT # heating/cooling rate
    return [dAdt, dBdt, dTdt]

# The model parameters
R = 8.314e-3 # kJ/mol*K
# Activation energy
Ea = 78 # kJ/mol
# Preexponential factor
k0 = 10**11.0 # L/mol*s
# Calculation time: 0 - 300s
t_range = np.linspace(0, 300)

# Initial conditions
A0 = 2 # mol/L
B0 = 0 # mol/L
T0 = 310 # K
ic = [A0, B0, T0] # initial conditions

# Case 1
# dT/dt = 0 isothermal
rT = 0
results = odeint(model, ic, t_range)
A = results[:,0]
B = results[:,1]
T = results[:,2]
plt.ylim(0,2)
plt.xlabel('Time')
plt.ylabel('[A], mol/L')
plt.plot(t,A)
plt.show()
plt.xlabel('Time, s')
plt.ylabel('T, K')
plt.plot(t,T)
plt.show()


```
