```python
# A <=> B
# A0 = 0.1
# K = 50


























# A = A0 - xi
# B = B0 + xi

import numpy as np
from scipy.optimize import fsolve

def model(X):
    xi = X
    A = A0 - xi
    B = B0 + xi
    eq1 = K*A - B # K - B/A
    return eq1

A0 = 0.1
B0 = 0
K = 5

guess = [0.01]

xi, = fsolve(model, guess)
A = A0 - xi
B = B0 + xi

print("Concentrations under equilibrium: [A]={:.3f}, [B]={:.3f}".format(A, B))
```

```python
# A <=> 2B
# A0 = 0.1
# K = 50





























# A = A0 - xi
# B = B0 + 2xi

import numpy as np
from scipy.optimize import fsolve

def model(X):
    xi = X
    A = A0 - xi
    B = B0 + 2*xi
    eq1 = K*A - B**2 # K - B**2/A
    return eq1

A0 = 0.1
B0 = 0
K = 5

guess = [0.01]

xi, = fsolve(model, guess)
A = A0 - xi
B = B0 + 2*xi

print("Concentrations under equilibrium: [A]={:.3f}, [B]={:.3f}".format(A, B))
```

```python
# A <=> B
# B <=> C
# A0 = 0.1
# K1 = 50
# K2 = 200



























import numpy as np
from scipy.optimize import fsolve

def model(X):
    xi1, xi2 = X
    A = A0 - xi1
    B = B0 + xi1 - xi2
    C = C0 + xi2
    
    eq1 = K1*A - B  # equivalent form of K1 - B/A
    eq2 = K2*A - C  # equivalent form of K2 - C/B
    
    return [eq1, eq2]

A0 = 0.1
B0 = 0
C0 = 0
D0 = 0
K1 = 50
K2 = 200

guess = [0.01,0.01]

xi1, xi2 = fsolve(model, guess)
A = A0 - xi1
B = B0 + xi1 - xi2
C = C0 + xi2
print("Concentrations under equilibrium: [A]={:.2e}, [B]={:.2e} [C]={:.2e}".format(A, B, C))
```
